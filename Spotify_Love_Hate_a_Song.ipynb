```python
c = DecisionTreeClassifier(min_samples_split=100)
```

```python
features = ["danceability", "duration_ms", "loudness", "speechiness", "valence", "energy", "key", "acousticness", "instrumentalness"]
```

```python
#training set
x_train = train[features] 
y_train = train["target"]

#test set
x_test = test[features] 
y_test = test["target"]
```

```python
x_train
```

```python
y_train
```

```python
#training the decision tree classifier

dt = c.fit(x_train,y_train)
```

```python
def show_tree(tree, features, path):
    f = io.StringIO()
    export_graphviz(tree, out_file=f, feature_names=features)
    pydotplus.graph_from_dot_data(f.getvalue()).write_png(path)
    img = imageio.imread(path)
    plt.rcParams["figure.figsize"] = (20,20)
    plt.imshow(img)
```

```python
show_tree(dt, features, 'dec_tree.png')
```

```python
y_pred= c.predict(x_test)
```

```python
y_pred
```

```python
#Calculating Accuracy on the test set (unseen data)

from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)*100
print(str(round(score,5)) + "%")
```
